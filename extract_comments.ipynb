```python
# select comments related to hygiene
```

```python
hygiene_keywords = [
    "cleanliness",
    "clean",
    "dirty",
    "sanitary",
    "hygienic",
    "disinfect",
    "germ-free",
    "sterile",
    "bacteria",
    "contamination",
    "health inspection",
    "filthy",
    "grime",
    "grease",
    "dust",
    "pest",
    "rodent",
    "insect",
    "mold",
    "mildew",
    "odor",
    "smell",
    "trash",
    "garbage",
    "litter",
    "stain",
    "unsanitary",
    "spoiled",
    "rotten",
    "food poisoning"
]
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
 nltk.download('omw-1.4')
```

```python
hygiene_rows = []
for index, row in date_df.iterrows():
    comment = row['comment']
    for keyword in hygiene_keywords:
        for word in nltk.word_tokenize(comment):
            lemma = lemmatizer.lemmatize(word.lower())
            if keyword == lemma:
                hygiene_rows.append(row)
    
```

```python
hygiene_df = pd.DataFrame(hygiene_rows)
```

```python
hygiene_df
```

```python
hygiene_df.to_csv("hygiene_df.csv")
```

```python
# calculate average for both datasets
```

```python
temp_rows1 = []
for index, row in hygiene_df.iterrows():
    rating_str = row['rating']
    rating_int = int(rating_str.split(' ')[0])
    row['rating'] = rating_int
    temp_rows1.append(row)
```

```python
new_hygiene_df = pd.DataFrame(temp_rows1)
```

```python
new_hygiene_df
```

```python
grouped_hygiene_df = new_hygiene_df.groupby('id')['rating'].mean()
```
