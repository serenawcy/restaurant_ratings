```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rating_df = pd.read_csv("../datasets/grouped_hygiene_df.csv")
```

```python
restaurant_df = pd.read_csv("../datasets/concat_restaurant_hygiene.csv")
```

```python
restaurant_df
```

```python
date_df = pd.read_csv("../datasets/grouped_date_df.csv")
```

```python
date_df
```

```python
df = pd.merge(rating_df, restaurant_df,  how='inner', left_on=['id'], right_on = ['id'])
```

```python
# Rewrite Correlation

hygiene_grades = df['GRADE'].copy()

revised_grade = []
for each in hygiene_grades:
  if each == 'A':
    revised_grade.append(0)
  elif each == "B":
    revised_grade.append(1)
  elif each == "C":
    revised_grade.append(2)
  else:
    revised_grade.append(3)
```

```python
df_corr = df.copy()
df_corr["GRADE_R"] = revised_grade

correlation = df_corr["GRADE_R"].corr(df_corr["rating_x"])
print("Correlation coefficient:", correlation)
```

```python
df_corr
```

```python
df2 = pd.merge(rating_df, date_df,  how='inner', left_on=['id'], right_on = ['id'])
```

```python
# Paired T-test
from scipy.stats import ttest_rel
t_statistic, p_value = ttest_rel(df2['rating_x'], df2['rating_y'])
print('t-statistic:', t_statistic)
print('p-value:', p_value)
```

```python
# Wilcoxon signed-rank test
from scipy.stats import wilcoxon
df2['difference'] = df2['rating_y'] - df2['rating_x']
statistic, p_value = wilcoxon(df2['difference'])
print('Wilcoxon statistic:', statistic)
print('p-value:', p_value)
```
