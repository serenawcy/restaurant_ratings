# Clean hygiene data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DOHMH_New_York_City_Restaurant_Inspection_Results.csv", encoding='latin-1')
```

```python
columns = ['CAMIS', 'DBA', 'BORO', 'ZIPCODE', 'Latitude', 'Longitude', 'CUISINE DESCRIPTION', 'SCORE', 'GRADE', 'GRADE DATE']
```

```python
new_df = df[columns]
new_df
```

```python
new_df = new_df[new_df['GRADE'].notna()]
df.shape, new_df.shape
```

```python
import datetime

temp_df = new_df.copy()
```

```python
temp_df['GRADE DATE'] = pd.to_datetime(temp_df['GRADE DATE'])
date = np.datetime64(datetime.date(2022, 1, 1))
data = temp_df[temp_df['GRADE DATE'] >= date]
```

```python
data.shape
```

```python
# remove rows where borough is unavailable
data = data[data['BORO']!="0"]
```

```python
data.shape
```

```python
# remove rows where latitude and longitude are unavailable
data = data[data['Latitude'].notna()]
data = data[data['Longitude'].notna()]
```

```python
data.shape
```

```python
# remove duplicate rows
data = data.drop_duplicates()
```

```python
data.shape
```

```python
# random sample 5000 rows
data = data.sample(n=5000)
```

```python
 # make sure indexes pair with number of rows
data = data.reset_index()
```

```python
data
```

```python
hygiene_data = data.drop(columns=['index'])
```

```python
hygiene_data
```

```python
hygiene_data.to_csv('hygiene_data.csv')
```

```python
def search_result(latitude, longitude, term):
    headers = {'Authorization': 'Bearer {}'.format(api_key)}
    url_params = {
        'latitude': latitude,
        'longitude': longitude,
        'term': term,
        'limit': 1
    }
    response=requests.get(url, headers=headers, params=url_params)
    return response.json()
```

```python
res_df = pd.DataFrame()
for index, row in hygiene_data.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    term = row['DBA']

    try:
        output_json = search_result(latitude, longitude, term)
        result_item = pd.DataFrame.from_dict(output_json["businesses"])
        res_df = res_df.append(result_item)
    except AttributeError:
        print("failed to retreive results for term".format(term))
```
