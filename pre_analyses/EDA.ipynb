```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("hygiene_data.csv")
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
sns.countplot(x = 'GRADE', data = df1) 
ax.set_title('Distribution of Hygiene Grade')
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
sns.countplot(x = 'BORO', data = df1) 
ax.set_title('Distribution of Borough')
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
sns.countplot(x = 'Cusine', data = df1) 
ax.set_title('Distribution of Borough')
```

```python
s = df1['CUISINE DESCRIPTION'].value_counts()
df1['CUISINE DESCRIPTION'] = np.where(df1['CUISINE DESCRIPTION'].isin(s.index[s < 50]), 'Other', df1['CUISINE DESCRIPTION'])
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
sns.countplot(x = 'CUISINE DESCRIPTION', data = df1) 
ax.set_title('Distribution of CUISINE DESCRIPTION')
ax.set(xlabel='Cuisine Type')
ax.tick_params(axis = 'x', rotation = 90)
```
